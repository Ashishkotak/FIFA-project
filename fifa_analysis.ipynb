```python
## FUNCTION TO GET WINNER AFTER STOPPAGE TIME i.e. IN EXTRA TIME OR PENALTIES

def get_winner_for_win_condition(win_condition):
    
    a= win_condition
    b=a.split()             # splitting the string so that we can access team name at 1st position i.e. at 0 index
    b.insert(-1,' ')       #in above case the win conditions where 
                            #there is blank it cant access the index hence we inserted dummy element
    winner_asper_wincondition= b[0]
    
    return winner_asper_wincondition
    
```

```python
matches['win_condition_winner']=matches.apply(lambda x: get_winner_for_win_condition(x['win_conditions']),axis=1)
```

```python
def get_winner(home_team,home_team_goals,away_team_goals,away_team):
    if home_team_goals > away_team_goals:
        winner = home_team
    elif home_team_goals == away_team_goals:
        winner = '--'
    else:
        winner=away_team
    return winner    
    

matches['winner']=matches.apply(lambda x:get_winner(x['home_team_name'],x['home_team_goals'],x['away_team_goals'],x['away_team_name']),axis=1)
```

```python
def get_team_ahead_at_half(home_team,half_time_home_goals,half_time_away_goals,away_team):
    if half_time_home_goals > half_time_away_goals:
        ahead= home_team
    elif half_time_home_goals == half_time_away_goals:
        ahead = 'scores_level'
    else:
        ahead=away_team
    return ahead    
```

```python
matches['team_ahead_at_half_time']=matches.apply(lambda x:get_team_ahead_at_half(x['home_team_name'],x['half_time_home_goals'],x['half_time_away_goals'],x['away_team_name']),axis=1)
```

```python
away_team_behind_at_half_and_won_game=pd.DataFrame(matches.groupby(['away_team_name','Year'])['away_team_behind_at_half_and_won_game'].sum())

away_team_behind_at_half_and_won_game.reset_index(inplace=True)

away_team_behind_at_half_and_won_game.rename(columns={'away_team_name':'team'},inplace=True)
```

```python
team_behind_at_halftime_and_won_game=pd.merge(home_team_behind_at_half_and_won_game,away_team_behind_at_half_and_won_game,on=['team','Year'])
```

```python
team_behind_at_halftime_and_won_game['matches_won_when_team_behind_at_halftime']=team_behind_at_halftime_and_won_game.home_team_behind_at_half_and_won_game+team_behind_at_halftime_and_won_game.away_team_behind_at_half_and_won_game
```
